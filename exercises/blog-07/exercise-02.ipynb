## Goal

Show long-range dependency fragility vs stability.

## ðŸ”¹ Task

Predict first token from last token.

Sequence length = 100.

Train tiny RNN and tiny Transformer.

Then inspect gradient norms for early positions.

## ðŸ§  Code (Core Part)

```python
import torch
import torch.nn as nn

device = "cuda" if torch.cuda.is_available() else "cpu"

seq_len = 100
dim = 32

x = torch.randn(1, seq_len, dim, requires_grad=True).to(device)

# RNN
rnn = nn.RNN(dim, dim, batch_first=True).to(device)
out, _ = rnn(x)
loss = out[:, -1].sum()
loss.backward()

print("RNN gradient norm first token:", x.grad[0, 0].norm())
```

Repeat with:

```python
x.grad.zero_()

attn = nn.MultiheadAttention(dim, 4, batch_first=True).to(device)
out, _ = attn(x, x, x)
loss = out[:, -1].sum()
loss.backward()

print("Attention gradient norm first token:", x.grad[0, 0].norm())
```

## ðŸ”Ž What They Observe

RNN gradient for early token shrinks more dramatically as sequence grows.

Attention gradient remains more stable.
