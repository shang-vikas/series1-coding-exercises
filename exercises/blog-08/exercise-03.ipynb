Now combine everything.

```python
import torch
import torch.nn as nn

class TinyTransformerLayer(nn.Module):
    def __init__(self, dim, heads=2):
        super().__init__()
        self.attn = nn.MultiheadAttention(dim, heads, batch_first=True)
        self.norm1 = nn.LayerNorm(dim)
        
        self.mlp = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.ReLU(),
            nn.Linear(dim*4, dim)
        )
        self.norm2 = nn.LayerNorm(dim)

    def forward(self, x):
        attn_out, _ = self.attn(x, x, x)
        x = self.norm1(x + attn_out)
        
        mlp_out = self.mlp(x)
        x = self.norm2(x + mlp_out)
        return x
```

Run:

```python
layer = TinyTransformerLayer(32)
x = torch.randn(2, 10, 32)
out = layer(x)

print("Input shape:", x.shape)
print("Output shape:", out.shape)
```

Now stack it 6 times.

```python
stacked = nn.Sequential(*[TinyTransformerLayer(32) for _ in range(6)])
out_stacked = stacked(x)

print("Stacked output shape:", out_stacked.shape)
```
