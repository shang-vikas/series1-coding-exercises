## Setup

```python
# Install required packages using the kernel's Python interpreter
import sys
import subprocess
import importlib

def install_if_missing(package, import_name=None):
    """Install package if it's not already installed."""
    if import_name is None:
        import_name = package

    try:
        importlib.import_module(import_name)
        print(f"‚úì {package} is already installed")
    except ImportError:
        print(f"Installing {package}....")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        print(f"‚úì {package} installed successfully")

# Install required packages
install_if_missing("numpy")
install_if_missing("matplotlib")
```

## üîπ 3Ô∏è‚É£ Gradient Descent Section

### üíª Coding Exercise 3 ‚Äî Real Gradient Descent

```python
import numpy as np

x = np.array([1, 2, 3])
y = np.array([2, 4, 6])

w = 0.0
lr = 0.01

for step in range(20):
    y_pred = w * x
    loss = np.mean((y_pred - y)**2)

    grad = 2 * np.mean((y_pred - y) * x)

    w -= lr * grad

    print(f"Step {step}: w={w:.3f}, loss={loss:.3f}")
```

We see:

- Loss decreasing
- Parameter moving
- No intelligence
- Just slope following.

### üíª Mini Exercise ‚Äî Gradient Sensitivity (Optional)

gradient is just sensitivity.

```python
def loss(w):
    x = np.array([1,2,3])
    y = np.array([2,4,6])
    return np.mean((w*x - y)**2)

w = 1.0
eps = 1e-5

numerical_grad = (loss(w+eps) - loss(w-eps)) / (2*eps)
print("Approx gradient:", numerical_grad)
```

This reinforces "nudge slightly and measure change."

No calculus required.

## üîπ 5Ô∏è‚É£ Overfitting Section

### üíª Coding Exercise 4 ‚Äî Overfitting Demo

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 10)
y = x + np.random.normal(0, 0.1, 10)

coeffs = np.polyfit(x, y, 9)

x_dense = np.linspace(0, 1, 100)
y_dense = np.polyval(coeffs, x_dense)

plt.scatter(x, y)
plt.plot(x_dense, y_dense)
plt.show()
```
