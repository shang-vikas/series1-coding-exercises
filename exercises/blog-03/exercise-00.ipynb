## Setup

````python
# Install required packages using the kernel's Python interpreter
import sys
import subprocess
import importlib

def install_if_missing(package, import_name=None):
    """Install package if it's not already installed."""
    if import_name is None:
        import_name = package

    try:
        importlib.import_module(import_name)
        print(f"âœ“ {package} is already installed")
    except ImportError:
        print(f"Installing {package}....")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        print(f"âœ“ {package} installed successfully")

# Install required packages
install_if_missing("numpy")
````

## Coding Exercise 1 â€” From Prediction to Loss


````python
import numpy as np

# simple dataset
x = np.array([1.0, 2.0, 3.0])
y_true = np.array([2.3, 4.1, 6.5])

# parameter
w = 0.5

# prediction
y_pred = w * x

# mean squared error
loss = np.mean((y_pred - y_true)**2)

print("Predictions:", y_pred)
print("Loss:", loss)
````

Now change `w` manually.

Observe loss change.

This makes error tangible.

### ðŸ’» Coding Exercise 2 â€” Accuracy vs Loss

````python
import numpy as np

y_true = np.array([1, 0, 1, 1])
y_pred = np.array([0.51, 0.49, 0.9, 0.6])

accuracy = np.mean((y_pred > 0.5) == y_true)
loss = np.mean((y_pred - y_true)**2)

print("Accuracy:", accuracy)
print("Loss:", loss)
````

Now change predictions to:

```python
y_pred = np.array([0.99, 0.01, 0.99, 0.99])
```

Accuracy stays same.

Loss drops.

You should feel why smooth signals matter.

````python
# Try with better predictions
y_true = np.array([1, 0, 1, 1])
y_pred_improved = np.array([0.99, 0.01, 0.99, 0.99])

accuracy_improved = np.mean((y_pred_improved > 0.5) == y_true)
loss_improved = np.mean((y_pred_improved - y_true)**2)

print("=== Improved Predictions ===")
print("Predictions:", y_pred_improved)
print("Accuracy:", accuracy_improved)
print("Loss:", loss_improved)
print("\nNote: Accuracy stays same (100%), but loss drops significantly! This means model is more confident in its predictions.")
print("This shows why smooth signals matter for gradient-based learning.")
````
